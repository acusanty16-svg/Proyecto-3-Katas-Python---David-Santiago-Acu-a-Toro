# Enunciado: Escribe un programa en Python que utilice condicionales para determinar el monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe:
    a. Solicitar al usuario el precio original de un artículo.
    b. Preguntar si tiene un cupón de descuento (respuesta sí o no).
    c. Si la respuesta es sí, solicitar el valor del cupón de descuento.
    d. Aplicar el descuento al precio original, siempre que el valor del cupón sea válido (mayor a cero).
    e. Mostrar el precio final de la compra, considerando o no el descuento.
    f. Usar estructuras de control de flujo (if, elif, else) para llevar a cabo las acciones.

### Ejemplo de compra con descuento

```python
def calcular_precio_final():
    """Calcula el precio final de una compra aplicando un descuento opcional.

    Solicita al usuario el precio original y si tiene un cupón de descuento.
    Si tiene cupón, pide el valor del descuento (1-100%) y lo aplica.
    Si no tiene cupón, muestra el precio original. Maneja errores de
    entrada con try/except y bucles de reintento.

    Returns:
        int or None: El precio final de la compra, o None si la opción
                     ingresada no es válida.
    """
    while True:
        try:
            precio_original = int(input("Introduce el monto que quieres pagar: "))
            cupon = input("¿Tienes cupon? Responde (si/no)").lower()
            if cupon == "no":
                print(f'No tienes cupon para aplicar descuento entonces tu compra es de: {precio_original}')
                return precio_original
            elif cupon == "si":
                descuento = int(input("Introduce el valor del cupon de descuento para continuar(1-100): "))
                while True:
                    if 0 <= descuento <= 100:
                        total = precio_original - (precio_original * descuento)/ 100
                        print(f"Precio original: {precio_original}")
                        print(f"Descuento aplicado: {descuento}%")
                        print(f"Precio final: {total}")
                        return total
                    else:
                        print("El valor introducido es incorrecto, intenta nuevamente")
                        descuento = int(input("Introduce el valor del cupon de descuento para continuar(1-100): "))
            else:
                print("Has introduce un dato incorrecto y no se puede continuar")
                return None
        except ValueError:
            print("Introduce un valor numerico valido")
    #Para el ultimo ejercicio decidi poner en practica casi todo lo que hemos visto en el modulo de python, con algunas excepciones claro esta, por ejemplo aqui no hay bucles, o entradas a diccionarios, pero creo que resume bastante bien todo el programa hecho hasta ahora, con if, elif y else aparte de manejar trys y excepts

calcular_precio_final()
```
