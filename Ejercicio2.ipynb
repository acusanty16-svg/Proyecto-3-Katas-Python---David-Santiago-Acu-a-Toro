## Segunda Kata

Enunciado: Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función map().

```python
numeros = [12, 45, 78, 3, 91, 26, 54, 67, 18, 39, 82, 7, 63, 49, 15, 88, 31, 72, 56, 24]

def doblar(numero):
    """doblar numero

    Args:
        numero (int): recibes un numero cualquiera

    Returns:
        int: doblas su valor al multiplicarlo por dos y lo formateas para que sea mas legible
    """
    return f'numero {numero} * 2 = {numero *2}'

def numeros_doble(args):
    """Doblar el valor de la lista

    Args:
        args (int): Al recibir una lista de numeros, hace un mapeo de cada numero con la funcion doblar y crea una nueva lista con los nuevos valores
    """
    resultado = list(map(doblar,args))
    for numero in resultado:
        print(numero)

numeros_doble(numeros)

# La idea de esta funcion es doblar el valor, pero era mejor hacerlo por separado y no unir las dos funciones en una sola, para tener mas legilibilidad, y aparte servira para explicar mejor el siguiente posible resultado
```

```python
def numeros_doble_sin_lista(*args):
    """El doble del valor de cada numero sin tener una lista previa

     Args:
        args (int): Al recibir una lista de numeros(sin estar previsualidos), hace un mapeo de cada numero con la funcion doblar y crea una nueva lista con los nuevos valores
    """
    resultado = list(map(doblar,args))
    for numero in resultado:
        print(numero)

numeros_doble_sin_lista(1,2,3,4,2,3,4,32,23,4,342,4,23)

# El concepto es el mismo, pero la diferencia radica en que aqui no necesitas de tener una lista previsualida con anterioridad, sino que solamente introduces manualmente los numeros y puedes ejecuar la funcion como mas convenga
```
