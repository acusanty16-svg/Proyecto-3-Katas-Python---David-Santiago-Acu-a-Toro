```python
def promedio ():
    """Calcula el promedio de una serie de números ingresados por el usuario.

    Solicita al usuario la cantidad de números a promediar, luego cada número
    individualmente validando que esté entre 0 y 10. Maneja errores de entrada
    con try/except anidados.

    Returns:
        float or None: El promedio calculado, o None si la cantidad es 0.
    """
    try:
        cantidad = int(input("Introduce la cantidad de numeros para hacer el promedio: "))
        if cantidad == 0:
            print("No se puede promediar 0")
            return None
        lista_numeros = []
        for i in range (cantidad):
            while True:
                try:
                    numero = int(input(f'Registra el numero {i+1} para continuar'))
                    if 0 <= numero <= 10:
                        break
                    print("Numero invalido (0-10)")
                except ValueError:
                    print("Introduce un numero valido")
            lista_numeros.append(numero)
        sumatoria = sum(lista_numeros)
        resultado = sumatoria / cantidad
        print(f'La media es: {resultado}')
    except ValueError:
        print("Introduce un numero valido para poder construir el promedio")
        
promedio()

# Bueno no voy a mentir jajajaj en este si que tuve un poco de ayuda porque al ejecutar el codigo empece a pensar en mil problemas que podrian pasar y en la cantidad de errores que se podrian generar entonces tuve que pedirle a mi fiel opencode que me ayudara a hacer varios try para que el codigo funcionase a la perfeccion
```
