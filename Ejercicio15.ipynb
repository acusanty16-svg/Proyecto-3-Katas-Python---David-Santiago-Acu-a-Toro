# Quinceava Kata

### Enunciado: Crea una función lambda que sume 3 a cada número de una lista dada.

### Ejemplo de la funcion lambda

```python
numeros = [1, 4, 7, 10, 13, 16, 19, 22, 25, 28]

resultado_suma = lambda numero: numero + 3
lista_nueva = list(map(resultado_suma, numeros))
print(lista_nueva)
# Este ejercicio si fue un poco mas sencillo, porque solamente he creado la funcion lambda iterando cada numero con su correspondiente suma 3, y despues creo una nueva lista donde voy a guardar los resultados y los imprimo por consola
```
