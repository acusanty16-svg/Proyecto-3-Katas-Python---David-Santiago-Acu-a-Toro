# Trigésimo quinta Kata

### Enunciado: Crea la clase UsuarioBanco
Representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta corriente.
Métodos: retirar_dinero, transferir_dinero, agregar_dinero.
Código a seguir:
Inicializar un usuario con nombre, saldo y un indicador (True o False) de cuenta corriente.
Implementar retirar_dinero para sustraer dinero del saldo, lanzando un error si no es posible.
Implementar transferir_dinero para transferir dinero desde otro usuario, lanzando un error en caso de fallo.
Implementar agregar_dinero para aumentar el saldo del usuario.
Caso de uso:
        a. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
        b. Agregar 20 unidades al saldo de Bob.
        c. Transferir 80 unidades de Bob a Alicia.
        d. Retirar 50 unidades del saldo de Alicia.

### Ejemplo de la clase usuario con POO

```python
class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente
        #Este ejercicio es tremendamente parecido al anterior asi que sera mejor comentarlo de igual forma. Para la primera parte, lo que he querido hacer es una clase como lo especificas dentro del enunciado, con nombre, saldo, y la cuenta_corriente

    def retirar_dinero(self, cantidad):
        try:
            if cantidad < self.saldo:
                self.saldo -= cantidad
                print(f'Has retirado {cantidad}. El saldo disponible es: {self.saldo}')
                return True
            if cantidad == self.saldo:
                self.saldo -= cantidad
                print(f'Macho te has quedado sin un duro al retirar {cantidad}')
                return True
            if cantidad > self.saldo:
                print("Saldo insuficiente")
                return False
        except (ValueError, TypeError):
            print("Valor introducido incorrectamente, intenta digitar un elemento correspondiente")
        #El primer metodo implementado es retirar dinero y tenemos 3 opciones, una donde restamos de la cantidad de la cuenta, la cantidad de dinero que queremos, la segunda es si quedamos en 0 y la tercera opcion es si el saldo es insuficiente, y todo lo he manejado dentro de un try
    
    def transferir_dinero(self, cantidad, cuenta_destino):
        try: 
            if not cuenta_destino.cuenta_corriente:
                print("Sin cuenta corriente no se puede hacer transferencia... !!¿Qué somos? ¿Animales?¡¡")
            else:
                if cantidad < self.saldo:
                    self.saldo -= cantidad
                    cuenta_destino.saldo += cantidad
                    print(f'Transferencia exitosa a la cuenta corriente con nombre: {cuenta_destino.nombre}')
                else:
                    print("Saldo insuficiente para hacer la transferencia")
        except (ValueError, TypeError):
            print("Valor introducido incorrectamente, intenta digitar un elemento correspondiente")
        #Para esta parte del ejercicio tengo tanto la cuenta destino como la cuenta de envio, donde restamos la cantidad de dinero de nuestra cuenta y luego esa misma cantidad la sumamos en la cuenta de destino, claro si no tienes cuenta corriente es imposible hacerlo y si no hay dinero el saldo es insuficiente. Y lo mismo, todo dentro de un try para que nadie se ponga gracioso a poner lo que quiera sino que aparezca el error
    
    def agregar_dinero (self, cantidad):
        try:
            total = cantidad + self.saldo
            print(f'Genial has hecho una transacción de {cantidad}. Tu nuevo saldo es de {total}')
            self.saldo += cantidad
        except (ValueError, TypeError):
            print("Valor introducido incorrectamente, intenta digitar un elemento correspondiente")
        #Y este es el metodo mas facil de todos, porque aqui simplemente introducidimos el valor a sumar y si alguien escribe un string o algo parecido pues saldra el error
            


alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)

bob.agregar_dinero(20)
bob.transferir_dinero(80,alicia)
alicia.retirar_dinero(50)
```

```python
def crear_usuario(nombre, saldo, cuenta_corriente):
    return {
        "nombre": nombre,
        "saldo": saldo,
        "cuenta_corriente": cuenta_corriente
    }
    #Este ejercicio es practicamente igual que en POO pero con diccionarios, porque pense que seria mejor verlo de dos formas. Para esta primera parte he hecho una funcion que crea un diccionario con los datos del usuario, como el nombre, el saldo y la cuenta corriente


def retirar_dinero(usuario, cantidad):
    try:
        if cantidad < usuario["saldo"]:
            usuario["saldo"] -= cantidad
            print(f'Has retirado {cantidad}. El saldo disponible es: {usuario["saldo"]}')
            return True
        if cantidad == usuario["saldo"]:
            usuario["saldo"] -= cantidad
            print(f'Macho te has quedado sin un duro al retirar {cantidad}')
            return True
        if cantidad > usuario["saldo"]:
            print("Saldo insuficiente")
            return False
    except (ValueError, TypeError):
        print("Valor introducido incorrectamente, intenta digitar un elemento correspondiente")
    #El metodo retirar dinero tiene 3 opciones al igual que en POO, una donde restamos de la cantidad de la cuenta, la cantidad de dinero que queremos, la segunda es si quedamos en 0 y la tercera opcion es si el saldo es insuficiente, y todo lo he manejado dentro de un try.


def transferir_dinero(usuario, cantidad, cuenta_destino):
    if not cuenta_destino["cuenta_corriente"]:
        print("Sin cuenta corriente no se puede hacer transferencia... !!¿Qué somos? ¿Animales?¡¡")
    else:
        if cantidad < usuario["saldo"]:
            usuario["saldo"] -= cantidad
            cuenta_destino["saldo"] += cantidad
            print(f'Transferencia exitosa a la cuenta corriente con nombre: {cuenta_destino["nombre"]}')
        else:
            print("Saldo insuficiente para hacer la transferencia")
    #Aqui he manejado tanto la cuenta destino como la cuenta de envio, para que de un lado reste el saldo actual y se lo envie en dos pasos a la cuenta destino y si el saldo es insuficiente saldra un warning avisando de ello


def agregar_dinero(usuario, cantidad):
    total = cantidad + usuario["saldo"]
    print(f'Genial has hecho una transacción de {cantidad}. Tu nuevo saldo es de {total}')
    usuario["saldo"] += cantidad
    #En este metodo lo que hago es sumar la cantidad que se quiere agregar al saldo actual del usuario y muestro el resultado


alicia = crear_usuario("Alicia", 100, True)
bob = crear_usuario("Bob", 50, True)

agregar_dinero(bob, 20)
transferir_dinero(bob, 80, alicia)
retirar_dinero(alicia, 50)
#Este ejercicio es igual al anterior ejemplo pero la diferencia es que manejo todo con diccionarios en vez de clases para que se pudiera ver que se puede hacer de dos formas el ejercicio
```
