### Enunciado: Concatena una lista de palabras. Usa la función reduce()

### Ejemplo sin tuplas

```python
from functools import reduce
palabras = ["That's", "what", "she", "said", "Bears", "Beets", "Battlestar", "Galactica", "Identity", "Theft", "is", "not", "a", "joke", "Jim", "MICHAEL", "Scott", "Dwight", "Schrute", "Prison", "Mike"]
def concatenar (lista):
    """Concatena todas las palabras de una lista en un solo string separado por espacios.

    Usa reduce() para acumular cada palabra con un espacio entre ellas.

    Args:
        lista (list): Lista de palabras (strings).

    Returns:
        str: String con todas las palabras concatenadas.
    """

    return reduce(lambda acumulador, palabra: acumulador +" "+palabra, lista).strip()
resultado = concatenar(palabras)
print(resultado)
# Este ejercicio es igual que al anterior, la unica diferencia es que fue necesario poner un .strip para los espacios que quedan en blanco pero de resto la mecanica del codigo es la misma, tener un acumulador mas la palabra que queremos-y, para que se entienda mejor la union de palabras he decidido poner un espacio en medio de ellas para que se lea mejor y sea mas entendible-y, con eso esta terminado
```

### Ejemplo con tuplas

```python
datos = [
    "Michael", "está",
    ("haciendo", "otro"),
    ("de los", "suyos"),
    "Dwight", "ama",
    ("la", "remolacha"),
    ("Jim", "mira"),
    "a", "cámara",
]

def concatenar_tuplas(*lista):
    """Extrae strings de una estructura mixta y los concatena en una sola frase.

    Recibe strings y tuplas como argumentos separados. Extrae los strings
    tanto del primer nivel como de dentro de las tuplas, los acumula en una
    lista y luego los concatena en un único string usando reduce().

    Args:
        *lista: Elementos de tipo str o tuple (que contienen str) pasados
                como argumentos individuales.

    Returns:
        str: Frase resultante de concatenar todas las palabras con espacios.
    """
    lista_nueva = []
    for elemento in lista:
        if isinstance(elemento, str):
            lista_nueva.append(elemento)
        if isinstance(elemento, tuple):
            lista_nueva.extend(list(filter(lambda palabra: isinstance(palabra, str), elemento)))
    return reduce(lambda acumulador, palabra: acumulador + " " + palabra, lista_nueva)

resultado = concatenar_tuplas(*datos)
print(resultado)
# Este ejercicio es muy parecido al ejercicio 20 donde he decidio mezclar tanto tuplas como strings para poder concatenarlos todos y asi practicar un poco mas y lo que he hecho es hacer un for de la todos los elementos dentro de la lista y separo los elementos que son strings y las tuplas en un instancia y al final devuelvo el resultado en reduce para obtener un solo resultado
```
