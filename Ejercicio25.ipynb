### Enunciado: Crea una función que cuente el número de caracteres en una cadena de texto dada.

### Ejemplo de lectura con cadena de texto. Reduce()

```python
from functools import reduce

texto = "I wish there was a way to know you're in the good old days before you've actually left them"

def contar_elementos_cadena_r(frase):
    """Cuenta el número de caracteres de una cadena de texto usando reduce().

    Itera sobre cada carácter de la frase con reduce(), sumando 1 al
    acumulador por cada letra. El acumulador inicia en 0.

    Args:
        frase (str): La cadena de texto a medir.

    Returns:
        int: Cantidad total de caracteres en la cadena.
    """

    return reduce(lambda acumulador, letra: acumulador + 1, frase, 0)
resultado = contar_elementos_cadena_r(texto)
print(f'El resultado total del siguiente texto: "{texto}" es: {resultado}')
# Este ejercicio me devuelve el numero de letras que existe en cada frase que introduzca, y he utilizado un reduce para reproducir el resultado, sin embargo la letra dentro del reduce no me interesa pero debo de declararla asi no la use sino no me va a funcionar el reduce solamente me interesa saber cuantas letras hay y con el acumulador es mas que suficiente
```

### Ejemplo de lectura con cadena texto. Bucle

```python
texto = "I wish there was a way to know you're in the good old days before you've actually left them"

def contar_elementos_cadena_b(frase):
    """Cuenta el número de caracteres de una cadena de texto usando un bucle.

    Itera sobre cada carácter de la frase con un bucle for, incrementando
    un contador en cada iteración.

    Args:
        frase (str): La cadena de texto a medir.

    Returns:
        int: Cantidad total de caracteres en la cadena.
    """

    contador = 0
    for caracter in frase:
        contador +=1
    return contador
resultado = contar_elementos_cadena_b(texto)
print(f'El resultado total del siguiente texto: "{texto}" es: {resultado}')
# Y el ejemplo del ejercicio de arriba es lo mismo pero la diferencia aqui es que hago un bucle para obtener el numero de caracteres o letras dentro de la frase
```

### Ejemplo de lectura en un diccionario

```python
datos = [
    "hola",
    42,
    ("a", "b", "c"),
    3.14,
    {"nombre": "Michael", "edad": 40},
    "mundo",
    7,
    (1, 2, 3, 4),
    99,
    True,
    {"serie": "The Office", "temporadas": 9},
    (10, 20),
    "Python",
    None,
    False,
    {"personaje": "Dwight", "trabajo": "Asistente del gerente regional"},
    (5, 6, 7, 8, 9)
]
def contar_elementos_diccionario(*lista):
    """Cuenta cuántos elementos de cada tipo hay en los argumentos recibidos.

    Itera sobre los argumentos, obtiene el nombre del tipo de cada uno
    con type().__name__, y acumula las frecuencias en un diccionario.

    Args:
        *lista: Elementos de tipos variados pasados como argumentos individuales.

    Returns:
        dict: Diccionario con tipo(str) -> cantidad(int).
    """
    contador={}
    for elemento in lista:
        tipo = type(elemento).__name__
        if tipo in contador:
            contador[tipo] += 1
        else:
            contador[tipo] = 1
    return contador
resultado = contar_elementos_diccionario(*datos)
print(f'En referencia a la siguiente lista: ')
for dato in datos:
    print(f'-{dato}')
print(f'Hemos podido obtener los siguientes resultados: {resultado}')
    
#En este ejercicio preferí usar un codigo muy parecido al primer ejercicio de las katas, en donde leo todos los elementos de una lista y dependiendo de su tipo asi mismo los guarda dentro de un nuevo diccionario con su nombre independiente y al final podremos ver la cantidad de cada uno. He decidido hacerlo de esta forma porque creo que es mas interesante
```
