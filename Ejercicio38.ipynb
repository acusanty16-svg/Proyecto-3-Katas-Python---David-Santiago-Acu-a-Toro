### Enunciado: Escribe un programa que determine qué calificación en texto tiene un alumno según su calificación numérica.
Reglas:
        0 - 69: insuficiente
        70 - 79: bien
        80 - 89: muy bien
        90 - 100: excelente

### Ejemplo de calificacion alumno

```python
def calificacion_texto():
    """Convierte una calificación numérica en su equivalente en texto.

    Solicita una nota (0-100) en un bucle continuo hasta que el usuario
    ingrese un valor válido. Según el rango, retorna "Insuficiente",
    "Bien", "Muy bien" o "Excelente".

    Returns:
        str: Calificación en texto correspondiente a la nota ingresada.
    """
    while True:
        try:
            nota = int(input("Introduce la calificacion numerica (0-100): "))
            if nota < 0 or nota > 100:
                print("Nota invalida. Debe ser entre 0 y 100")
                continue
            if nota <= 69:
                return "Insuficiente"
            elif nota <= 79:
                return "Bien"
            elif nota <= 89:
                return "Muy bien"
            else:
                return "Excelente"
        except ValueError:
            print("Introduce un numero valido para continuar")
    #Es casi el mismo ejercicio que hemos hecho anteriormente, y la explicacion es la misma, dada una calificacion recibiras un mensaje correspondiente 

print(calificacion_texto())
```
