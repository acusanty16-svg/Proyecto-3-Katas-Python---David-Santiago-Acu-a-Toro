# Decimosexta Kata

### Enunciado: Escribe una función que tome una cadena de texto y un número entero n como parámetros y devuelva una lista de todas las palabras que sean más largas que n. Usa la función filter().

### Ejemplo sin input

```python
texto = "El que tenga el poder de derrotar al Señor Tenebroso se acerca, nacerá de aquellos que lo han desafiado tres veces"
def buscar_palabra(frase, numero):
    """Filtra las palabras de una frase que superan una longitud determinada.

    Separa la frase en palabras individuales y utiliza filter() con una
    lambda para conservar solo aquellas cuya longitud sea mayor que el
    número indicado.

    Args:
        frase (str): La cadena de texto a analizar.
        numero (int): Longitud mínima para considerar una palabra.

    Returns:
        list: Lista de palabras con más caracteres que el número indicado.
    """
    palabras = frase.split()
    return list(filter(lambda palabra: len(palabra) > numero, palabras))
resultado = buscar_palabra(texto, 6)
print(resultado)
# Para este ejercicio lo que priorizado es crear un lista que me devuelva las palabras que tengan un numero mayor de caracteres al introducido
```
