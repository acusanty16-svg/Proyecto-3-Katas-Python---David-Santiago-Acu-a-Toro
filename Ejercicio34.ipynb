### Enunciado: Crea la clase Arbol
Define un árbol genérico con un tronco y ramas como atributos.
Métodos disponibles: crecer_tronco, nueva_rama, crecer_ramas, quitar_rama, info_arbol.
Código a seguir:
Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
Implementar el método quitar_rama para eliminar una rama en una posición específica.
Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y sus longitudes.
Caso de uso:
        a. Crear un árbol.
        b. Hacer crecer el tronco una unidad.
        c. Añadir una nueva rama.
        d. Hacer crecer todas las ramas una unidad.
        e. Añadir dos nuevas ramas.
        f. Retirar la rama situada en la posición 2.
        g. Obtener información sobre el árbol.

### Ejemplo de la clase arbol con POO

```python
# Tuve un poco de dudas en este ejercicio, porque al pedir la clase arbol, me imagine que querias que hicieramos el trabajo con programacion orientada a objetos, de igual forma lo he intentado hacer con diccionarios. La informacion de este ejercicio la suminstra con mi amigo Youtube jejejej
from functools import reduce
class Arbol:
    def __init__(self):
        self.tronco = 1
        self.ramas = []
    # Para este ejercicio si comentaré todo porque todo la informacion la he sacado adicional, entonces creo que será mejor de este forma, para empezar creamos la clase arbol y para su inicalizacion le decimos que tiene un tronco y una lista con ramas vacia por el momento

    def crecer_tronco(self):
        self.tronco += 1
        print(f'El tronco ha crecido una unidad mas, ahora es de {self.tronco}')
    #Cada vez que se use el metodo, crecerá una unidad mas

    def aniadir_rama(self):
        self.ramas.append(1)
        print(f'El arbol ahora tiene una rama mas, su cantidad es de {len(self.ramas)}')
    # En este metodo se añade una rama nueva y se pone por defecto la unidad 1 como base y a medida que va creciendo el arbol asimismo su longitud

    def crecer_ramas(self):
        for i in range(len(self.ramas)): 
            self.ramas[i] += 1
        print(f"La cantidad de ramas es de {len(self.ramas)} y la lista de cada una con su nueva longitud es: {self.ramas}")
    # Aqui decimos que por cada rama se sumara un 1 a su longitud, entonces si ha habido ramas anteriores estas creceran, pero si una nueva llega entonces no se sumara sino que su base sera de 1

    def aniadir_dos_ramas(self):
        self.ramas.append(1)
        self.ramas.append(1)
        print(f"Se han añadido dos ramas mas el nuevo total es {len(self.ramas)}")
        print(f"Y su correspondiente lista es: {self.ramas}")
    # Mismo ejercicio que el anterior pero la diferencia es que aqui añado dos ramas en vez de una

    def retirar_rama(self, posicion):
        for i in self.ramas:
            if posicion == i:
                self.ramas.pop(posicion)
                print(f"Se ha podado la rama en la posicion {posicion} y ahora el arbol tiene {self.ramas} en total :(")
                return
            if posicion <= 0:
                print("Error al cortar la rama, nos quedamos sin gasolina")
                return
    
    def info_arbol(self):
        print(f"Longitud del tronco: {self.tronco}")
        print(f"Número de ramas: {len(self.ramas)}")
        print(f"Longitudes de cada rama: {self.ramas}")
        if self.ramas:
            print(f"Longitud total de ramas: {sum(self.ramas)}")
        

arbol = Arbol()
arbol.crecer_tronco()

arbol.aniadir_rama()
arbol.aniadir_rama()
arbol.crecer_ramas()
arbol.crecer_ramas()
arbol.aniadir_rama()
arbol.crecer_ramas()
arbol.aniadir_rama()
arbol.crecer_ramas()

arbol.aniadir_dos_ramas()

arbol.retirar_rama(5)
arbol.retirar_rama(-50)

arbol.info_arbol()
```
