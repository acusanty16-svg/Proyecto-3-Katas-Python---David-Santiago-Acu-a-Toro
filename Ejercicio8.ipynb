### Enunciado: Escribe un programa que pida al usuario dos números e intente dividirlos. Si el usuario ingresa un valor no numérico o intenta dividir por cero, maneja esas excepciones de manera adecuada y muestra un mensaje indicando si la división fue exitosa o no.

### Primer ejemplo sin try 

```python
def dividir_dos_numeros(numero1, numero2):
    """dividir dos numeros sin try

    Args:
        numero1 (int or float): si el parametro introducido no es correcto, entonces saldra una alerta diciendo que debe ser un float o un integer 
        numero2 (int or float): si el parametro introducido no es correcto, entonces saldra una alerta diciendo que debe ser un float o un integer y diferente de 0

    Returns:
        float: devuelve en un float el resultado de la division realizada
    """
    if not isinstance(numero1, (int, float)):
        print("Introduce un número para poder dividir")
        return False
    if numero2 == 0:
        print("Introduce un valor diferente de 0 para continuar")
        return False
    resultado = numero1 / numero2
    print(f'{numero1} / {numero2} = {resultado}')
    return resultado

dividir_dos_numeros(5,5)
# Para este ejercicio he decidido realizarlo de dos formas, una sin try catch para captar los errores y la siguiente forma de hacerlo si que aplicare los try, porque hay mas margen de juego, quiero decir que esta forma no está mal del todo, pero si introduzco un string con numero tendria que dejar mucha mas linea de codigo para leerlo
```

### Segundo ejemplo con try

```python
def dividir_dos_numeros_try():
    """dividir dos numeros con try

    Args:
        numero1 (int or float): si el parametro introducido no es correcto, entonces saldra una alerta diciendo que debe ser un float o un integer 
        numero2 (int or float): si el parametro introducido no es correcto, entonces saldra una alerta diciendo que debe ser un float o un integer y diferente de 0

    Returns:
        float: devuelve en un float el resultado de la division realizada
    """
    try: 
        numero1 = int(input("Introduce el primer numero: "))
        numero2 = int(input("Introduce el segundo numero: "))
        resultado = numero1/numero2
    except ValueError:
        print("Escribir un numero adecuado es obligatorio")
    except ZeroDivisionError:
        print("Imposible dividir por zero crack")
    else:
        print(f'{numero1} / {numero2} = {resultado}')
        return resultado

dividir_dos_numeros_try()
# El mismo ejercicio, pero para esta ocasion si que he decidido usar el input porque da mas juego con las try, pero podria hacerse sin ello, simplemente declarando la firma diferente y pasando los resultados en el ultimo return. 
```
