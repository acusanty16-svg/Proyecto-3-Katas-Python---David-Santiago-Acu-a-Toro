### Enunciado: Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5). La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

### Primer ejemplo con una sola lista

```python
def media (notas):
    """Calcula la media de tus notas

    Args:
        notas (list): Introduce todas tus notas en un rango de 1-10

    Returns:
        tuple: devuelve una tupla que contiene tanto mensajes por consola con el respectivo estado, como tambien la informacion de la media
    """
    sumatorio = sum(notas)
    cantidad = len(notas)
    resultado = round(sumatorio / cantidad, 2)
    nota_aprobado = 5
    if 0 <= resultado <= 10:
        if resultado >= nota_aprobado:
            estado="aprobado"
            print(f'!!Muy bien¡¡ Eres toda una maquina y tu estado es: {estado} con {resultado} de media, Crack!!')
        else:
            estado= "suspenso"
            print(f'Suerte para la proxima hijo, a estudiar un poco mas, tu estado es {estado} y te has quedado con {resultado} de media... esperaba mas de ti')
    else:
        estado = "invalido"
        print(f'Rango invalido...')
    return (resultado, estado)

notas_santiago = [10,2,10,2,5,6,4,3,2,10,1,4,5]

resultado = media(notas_santiago)
print(type(resultado))
## Para este primer ejemplo queria que un solo estudiante pudiese ver su media con las notas correspondientes, entonces cree una variable que suma todos los numeros, luego una que cuenta cuantos elementos hay y al final tuve que redondear el resultado para solo tener dos decimales y devolverlos con el correspondiente mensaje y estado
```
