### Enunciado: Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo", "circulo" o "triangulo") y datos (una tupla con los datos necesarios para calcular el área de la figura).

### Ejemplo de figura

```python
def calcular_area(figura, datos):
    try:
        if figura == "rectangulo":
            base, altura = datos
            area = base * altura
            print(f'Has introducido: {datos}\nEl area del rectangulo es: {area}')
            return area
        elif figura == "circulo":
            radio = datos[0]
            area = 3.1416 * radio ** 2
            print(f'Has introducido: {datos}\nEl area del circulo es: {area}')
            return area
        elif figura == "triangulo":
            base, altura = datos
            area = (base * altura) / 2
            print(f'Has introducido: {datos}\nEl area del triangulo es: {area}')
            return area
        else:
            print("Figura no valida")
            return None
    except (ValueError, TypeError, IndexError):
        print("Error: introduce datos validos para la figura seleccionada")
        return None
    #Para este penúltimo ejercicio decidí calcular el area como nos dijistes con dos firmas, una para la figura y la otra para los datos que vamos a poner dentro para aveiguar el area, pero esto tenia que ponerlo dentro de un try, porque si no cualquier notas podria venir y poner calcular_area("rectangulo", ("hola")) y apareceria el error pero rompiendo el programa entonces mejor capturarlo y lo demas es como los ejercicio anteriores con un if dependiendo de que escogas

calcular_area("rectangulo", (5, 3))
calcular_area("rectangulo", (5, 3, 2))
calcular_area("circulo", ("hola",))
calcular_area("triangulo", (4, "tres"))

```
