### Enunciado: Crea una función lambda que calcule el resto de la división entre dos números dados.

### Ejemplo con tuplas

```python
datos = [(10, 3), (15, 4), (20, 7), (9, 2), (100, 11), (8, 5)]
def resultado_division(division):
    """Calcula el módulo (resto) de cada par de números en una lista de tuplas.

    Usa map() con una lambda que desempaqueta cada tupla (a, b) y
    devuelve a % b.

    Args:
        division (list): Lista de tuplas (dividendo, divisor).

    Returns:
        list: Lista con los restos de cada división.
    """
    return list(map(lambda par: par[0] % par[1], division))
resultado = resultado_division(datos)
for i, par in enumerate(datos):
    print(f'El siguiente par de numeros: {par}. Contiene el siguiente modulo: {resultado[i]}')
# Para este ejercicio opte por un enfoque un poco mas sencillo, si se tiene una serie de elementos en tupla simplemente se crea una funcion que haga un mapeo a la nueva lista que nos permite sacar el modulo de cada elemento y al final para cada uno de los elementos dentro de la tupla reproducimos el resultado en un bucle con su correspondiente modulo
```

### Ejemplo con input

```python
def resultado_division_input():
    """Solicita dos números al usuario y calcula el módulo (resto) de la división.

    Pide ambos números en un bucle hasta que se ingrese un divisor distinto
    de cero. Maneja errores de ValueError si la entrada no es numérica.

    Returns:
        int: El resto de dividir numero_1 entre numero_2.
    """
    while True:
        try:
            numero_1 = int(input("Introduce el primer numero a dividir: "))
            numero_2 = int(input("Introduce el segundo numero a dividir: "))
            if numero_2 == 0:
                print("Introduce un numero diferente a 0 para continuar")
            else:
                resultado = numero_1 % numero_2
                print(f'{numero_1} % {numero_2} = {resultado}')
                return resultado
        except ValueError:
            print("Introduce un numero valido para continuar")
    
resultado = resultado_division_input()
# Quise hacer el mismo ejercicio pero poniendo los numeros dentro de un input para que reprodujese el resultado de la misma manera que el otro pero un poco mas interactivo
```
