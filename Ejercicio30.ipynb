# Trigésima kata

### Enunciado: Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

### Ejemplo con una lista de tuplas

```python
anagramas = [("listen", "silent"),("hello", "world"),("amor", "roma"),("python", "java"),("desorden", "seronden"),]
def son_anagramas(lista):
    """Determina si cada par de palabras en una lista de tuplas son anagramas.

    Para cada tupla (palabra1, palabra2) compara las versiones ordenadas
    de ambas palabras con sorted(). Si son iguales, son anagramas.

    Args:
        lista (list): Lista de tuplas (str, str) con pares de palabras.

    Returns:
        None: La función imprime los resultados pero no devuelve ningún valor.
    """
    for palabra1, palabra2 in lista:
        if sorted(palabra1) == sorted(palabra2):
            print(f'"{palabra1}" y "{palabra2}" → Son anagramas')
        else:
            print(f'"{palabra1}" y "{palabra2}" → No son anagramas')
son_anagramas(anagramas)
# Para este ejercicio use sorted porque me ordena alfabeticamente los elementos de la lista entonces para este ejercicio es super esencial y lo que hice despues fue simplemente comparar y ponerlo por consola, si no queria hacerlo de esta forma y simplemente devolver el resultado en una lista pues podria hacer en un bucle cada resultado e imprimirlo por consola tambien
```
