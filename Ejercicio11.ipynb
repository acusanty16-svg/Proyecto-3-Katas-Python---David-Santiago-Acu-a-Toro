### Enunciado: Escribe un programa que pida al usuario que introduzca su edad. Si el usuario ingresa un valor no numérico o un valor fuera del rango esperado (por ejemplo, menor que 0 o mayor que 120), maneja las excepciones adecuadamente.

# Único ejemplo de la selección de edad

```python
def seleccion_edad():
    """Solicita y valida la edad del usuario.

    Pide al usuario que introduzca su edad en un bucle continuo hasta
    que ingrese un valor numérico dentro del rango 0-100. Maneja errores
    de tipo ValueError cuando la entrada no es un número.

    Returns:
        None: La función imprime el resultado pero no devuelve ningún valor.
    """
    while True:
        try:
            edad = int(input("Introduce tu edad: "))
            if 0 <= edad <=100:
                print(f'La edad "{edad}" introducida es correcta. Gracias por ser una persona real')
                break
            else:
                print("Introduce una edad valida para continuar (0-100). Intenta de nuevo")
        except ValueError:
            print("Debes introduce un numero para poder continuar")

seleccion_edad()

# Este ejercicio es un poco parecido al anterior y me sirvio mucho lo que aprendi en el anterior ejercicio porque no solamente aprendi a manejar las excepciones sino a que el mismo programa me ayuda reproducir el juego mucho mas fluido. Porque gracias al while True, cada vez que se ejecuta el resultado verdadero hace el break y el bucle termina, pero mientras sea False se repetira una y otra vez hasta que introduzcan el valor correspondiente
```
