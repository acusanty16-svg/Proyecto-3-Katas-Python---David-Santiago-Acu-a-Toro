### Enunciado: Para una lista con elementos de tipo integer y string, obtén una nueva lista solo con los valores int. Usa la función filter().

### Ejemplo sin *args

```python
lista_mixta = [1, "gato", 5, "perro", 3, "pájaro", 8, "elefante", 2, "tortuga", 7, "león"]
def obtener_numeros(lista):
    """Filtra los elementos de tipo entero de una lista mixta.

    Usa filter() con isinstance() para conservar solo los elementos
    que son de tipo int, descartando strings y otros tipos.

    Args:
        lista (list): Lista con elementos de tipos mixtos (int, str, etc.).

    Returns:
        list: Nueva lista conteniendo solo los valores enteros.
    """

    return list(filter(lambda numero: isinstance(numero, int), lista ))

resultado = obtener_numeros(lista_mixta)
print(f'Teniendo la siguiente lista mixta: {lista_mixta}\nHemos podido encontrar los siguientes numeros: {resultado}')
# Para este ejercicio lo que he querido hacer es tener un serie de strings e ints para poder encontrar solamente los numeros con un isinstance que lee el "numero" a iterar y el parametro que necesito que es el integer y devuelve la lista nueva
```

### Ejemplo con *args 

```python
lista_mixta = [
    1,
    "gato",
    (5, 3),
    "perro",
    8,
    ("pájaro", 2),
    "elefante",
    7,
    "león",
    (4, 9, 1)
]

def obtener_numero (*args):
    """Extrae todos los números enteros de los argumentos, incluyendo los anidados en tuplas.

    Itera sobre los argumentos recibidos. Si un argumento es un entero,
    lo agrega directamente. Si es una tupla, filtra sus elementos enteros
    con filter() y los agrega a la lista con extend().

    Args:
        *args: Elementos de tipos mixtos (int, str, tuple, etc.).

    Returns:
        list: Lista con todos los enteros encontrados.
    """
    lista_nueva=[]
    for elemento in args:
        if isinstance(elemento, int):
            lista_nueva.append(elemento)
        elif isinstance(elemento, tuple):
            lista_nueva.extend(list(filter(lambda numero: isinstance(numero, int), elemento)))
    return lista_nueva

resultado = obtener_numero(*lista_mixta)
print(f'Tras realizar una dura investigacion dentro de: {lista_mixta}\nHemos encontrado estos numeros dentro de ella: {resultado}')
# Para este ejercicio si necesite ayuda extra dentro del bucle, porque no comprendia muy bien como anexar o almacenar los numeros dentro de las tuplas para que no se rompiera y como son muchos elementos era necesario hacer un * entonces lo que decidi hacer fue un extend que añadiera dichos elementos y lo demas del codigo es lo mismo que el anterior elemento pero con una lista que añadi hijos con el condicional de integer
```
