### Enunciado: Genera una función que, al recibir una frase, devuelva una lista con la longitud de cada palabra. Usa la función map()

### Ejemplo 1 con input

```python
def construccion_frase():
    """Calcula la longitud de cada palabra en una frase ingresada por el usuario.

    Solicita una frase al usuario, separa las palabras y calcula cuántos
    caracteres tiene cada una usando map(len). Muestra la frase original
    y la lista de longitudes.

    Returns:
        None: La función imprime los resultados pero no devuelve ningún valor.
    """
    while True:
        try:
            frase = str(input("Introduce la palabra para saber cuantos caracteres tiene: "))
            palabra = frase.split()
            longitudes = list(map(len,palabra))
            print(f'La frase introducida ha sido: {frase}')
            print(f'Y la longitud de cada palabra es: {longitudes}')
            break
        except ValueError:
            #Segun lo que tengo entendido este codigo nunca va a llegar hasta aqui porque siempre ejecutara un string, pero bueno lo dejo por si acaso, no pasa nada
            print("Error al digitalizar, intentalo mas tarde")

construccion_frase()
# Para este trabajo simplemente he realizado un input, pero si se cambiara la firma se podria poner lo mismo pero leeria la lista o el string dependiendo lo que se quiera hacer y lo que se quiera leer
```

### Ejemplo 2 con lista

```python
lista_strings = ["Hola me llamo santiago", "Mi nombre es carla", "Y Yo", "Yo soy batman"]
def contar_frase(frase):
    """Cuenta la cantidad de palabras en una frase.

    Args:
        frase (str): La frase a evaluar.

    Returns:
        int: Número de palabras en la frase.
    """

    return len(frase.split())
def palabras_por_frase(*palabras):
    """Calcula cuántas palabras tiene cada frase recibida como argumento.

    Args:
        *palabras (str): Una o más frases pasadas como argumentos separados.

    Returns:
        list: Lista con el número de palabras de cada frase.
    """

    resultado = list(map(contar_frase,palabras))
    for i in range(len(palabras)):
        print(f'La frase "{palabras[i]}" tiene {resultado[i]} palabras')

palabras_por_frase(*lista_strings)

# Para este ejemplo lo que decidi hacer fue una funcion que cuente las frases pero que les quite el espacio y ya en la segunda que lea cada una de las frases en un rango y luego me muestre por consola cada una de ellas y su correspondiente longitud de palabras
```

### Ejemplo con string

```python
frase = "Hola, vengo de Mordor"
def leer_frase(frase):
    """Muestra la longitud de cada palabra en una frase.

    Separa la frase en palabras, calcula cuántos caracteres tiene cada una
    usando map(len, frase.split()), e imprime tanto la frase original como
    la lista de longitudes.

    Args:
        frase (str): La frase a analizar.

    Returns:
        None: La función imprime el resultado pero no devuelve ningún valor.
    """
    longitud_nueva_frase = list(map(len,frase.split()))
    print(f'La frase "{frase}" tiene: {longitud_nueva_frase} caracteres')

leer_frase(frase)
# Para este caso lo que he decidido hacer es hacer un mapeo en una lista nueva para poder contar y guardar el numero de caracteres dentro de la frase introducida
```
