# Vigésimo primera Kata

### Enunciado: Crea una función que calcule el cubo de un número dado mediante una función lambda.

### Ejemplo con Map

```python
numeros = [2, 5, 3, 7, 4, 9, 6, 8, 10, 1]
def numero_al_cubo(lista):
    """Calcula el cubo de cada número en una lista.

    Aplica map() con una lambda que eleva cada número al cubo (x * x * x).

    Args:
        lista (list): Lista de números enteros.

    Returns:
        list: Nueva lista con los cubos de los números originales.
    """

    return list(map(lambda numero: numero * numero * numero, lista))
resultado = numero_al_cubo(numeros)
print(resultado)
# Este ejemplo es casi igual que hacer un mapeo como en algun ejercicio anterior pero aqui he iterado la multiplicacion de su numero al cubo y lo devuelvo con en una lista
```

### Ejemplo sin Map

```python
numeros = [2, 5, 3, 7, 4, 9, 6, 8, 10, 1]
cubo = lambda numero: numero * numero * numero
lista_nueva = []
for numero in numeros:
    lista_nueva.append(cubo(numero))
print(f'Los numeros son: {numeros}\nSus cubos son: {lista_nueva}')
# La diferencia aqui es que he puesto la lambda como funcion unica y ya la aplico dentro del for y hago un append para guardar el resultado y lo imprimo por consola
```
