### Enunciado: Escribe una función que calcule el factorial de un número de manera recursiva

### Primer ejemplo sin input

```python
def factorial (numero):
    """Funcion para conocer factorial de un numero

    Args:
        numero (int): Introduce el numero que quieras para calcular su factorial

    Returns:
        int: Devulve el numero factorial con su correspondiente mensaje
    """
    if numero <= 1:
        return 1
    else:
        return numero * factorial(numero - 1)

numero = 7
resultado = factorial(numero)
print (f'El numero {numero} tiene un factorial de: {resultado}')

# Para este ejercicio he puesto una funcion que se llame a si misma para que pueda devolver el argumento correspondiente de cada numero multiplicado por sus anteriores. lo unico donde he tenido errores o no he sabido como hacerlo es la parte final, porque quise que el numero introducido dentro de la fucion por ejemplo el 5 fuese resultado = factorial(5) y que en la siguiente linea de codigo donde hago el print pudiese sacar un .get o algo para obtener el resultado que tengo, pero al no saber como hacerlo dentro de una funcion lo he puesto manualmente. Pero he dejado un opcion con input
```
