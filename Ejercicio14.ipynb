### Enunciado: Crea una función que retorne las palabras de una lista que comiencen con una letra en específico. Usa la función filter().

### Ejemplo con una lista de palabras

```python
palabras = ["gato", "perro", "gallina", "pato", "gorrión", "paloma", "gaviota", "pavo", "gacela", "puma"]

def buscar_palabra(palabra, letra):
    """Verifica si una palabra comienza con una letra específica.

    Args:
        palabra (str): La palabra a evaluar.
        letra (str): La letra inicial a buscar.

    Returns:
        bool: True si la palabra empieza con la letra indicada, False en caso contrario.
    """
    return palabra.startswith(letra)

def iterar_palabras(lista, letra):
    """Filtra las palabras de una lista que comienzan con una letra específica.

    Args:
        lista (list): Lista de palabras a filtrar.
        letra (str): Letra inicial por la que filtrar.

    Returns:
        list: Nueva lista con las palabras que comienzan con la letra indicada.
    """
    return list(filter(lambda palabra: buscar_palabra(palabra, letra), lista))
iterar_palabras(palabras, "p")
# Para este ejercicio he tenido que hacer dos funciones para que fuera un poco mas entendible, lo que he hecho es buscar la palabra por un lado y hacer un startswith para que me devuelva un booleano encontrando la palabra, despues en la otra funcion lo que he decidido hacer es recibir la lista, con la letra seleccionada y seguidamente hacer un filtrado con lambda que me permita tener una nueva lista con las palabras de la letra seleccionada
```

### Ejemplo extra con tuplas dentro de la lista

```python
datos = [
    ("gato", 3),
    ("perro", 7),
    ("gallina", 2),
    ("pato", 1),
    ("gorrión", 4),
    ("paloma", 5),
    ("gaviota", 6),
    ("pavo", 3),
    ("gacela", 8),
    ("puma", 10)
]

def buscar_palabra_tupla(tupla, letra):
    """Verifica si el nombre del animal en una tupla comienza con una letra específica.

    Args:
        tupla (tuple): Tupla (nombre_animal, edad).
        letra (str): Letra inicial a buscar.

    Returns:
        bool: True si el nombre empieza con la letra indicada, False en caso contrario.
    """

    return tupla[0].startswith(letra)

def iterar_palabras_tupla(lista, letra):
    """Filtra una lista de tuplas (animal, edad) cuyo nombre comience con una letra dada.

    Args:
        lista (list): Lista de tuplas (nombre, edad).
        letra (str): Letra inicial para filtrar.

    Returns:
        list: Nueva lista con las tuplas que cumplen la condición.
    """

    return list(filter(lambda tupla: buscar_palabra_tupla(tupla, letra), lista))
iterar_palabras_tupla(datos, "p")
# Para este ejercicio he tenido que hacer dos funciones para que fuera más entendible. primero busco la palabra con startswith para que devuelva un booleano indicando si el nombre del animal comienza con la letra deseada. Después, en la otra función, recibo la lista de tuplas y la letra seleccionada, y filtro con filter() y lambda para obtener una nueva lista con solo las tuplas de la letra indicada.
```
