### Enunciado: Dada una lista numérica, obtén el producto total de los valores. Usa la función reduce().

### Ejemplo de producto total

```python
from functools import reduce
numeros = [2, 3, 4, 5, 6]
def producto_total(lista):
    """Calcula el producto total de todos los números en una lista.

    Usa reduce() con una lambda que multiplica acumulativamente cada
    elemento para obtener el producto total de la lista.

    Args:
        lista (list): Lista de números enteros.

    Returns:
        int: Producto total de todos los números de la lista.
    """

    return reduce(lambda acumulador, numero: acumulador * numero, lista)
resultado = producto_total(numeros)
print(f'La lista de numeros es: {numeros}\nSu correspondiente producto total es: {resultado}')
# Muy parecido al ejercicio 17 que hemos hecho con anterioridad, la unica diferencia es que aqui necesito multiplicar el numero siguiente de la lista y no tengo que replicar el numero dado en la lista.
```
