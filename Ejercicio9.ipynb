# Novena Kata

### Enunciado: Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]. Usa la función filter().

### Ejemplo 1 y 2 de mascotas prohibidas

```python
mascotas = ["Perro", "Gato", "Mapache", "Loro", "Serpiente Pitón","Hámster", "Cocodrilo", "Conejo", "Oso", "Tortuga","Hurón", "Erizo", "Camaleón", "Tarántula", "Serpiente Coral","Lémur", "Cebra"]
prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]

def exlcuir_mascotas(mascota):
    """Filtra mascotas no prohibidas.

    Args:
        mascota (str): Nombre de la mascota a evaluar.

    Returns:
        bool: True si la mascota NO está en la lista de prohibidas,
              False si sí está prohibida.
    """
    if mascota in prohibidas:
        return False
    else:
        return True


def mascotas_prohibidas(mascota):
    """Filtra solo las mascotas prohibidas.

    Args:
        mascota (str): Nombre de la mascota a evaluar.

    Returns:
        bool: True si la mascota SÍ está en la lista de prohibidas,
              False si no lo está.
    """
    if mascota in prohibidas:
        return True
    else:
        return False


lista_nueva_excluir = list(filter(exlcuir_mascotas, mascotas))
print("Un resumen de las mascotas que no son prohibidas")
for indice, mascota in enumerate(lista_nueva_excluir, start=1):
    print(f'{indice} - {mascota}')
# Quise implementar aqui dos partes del trabajo aunque solamente pediste uno, igualmente queria agregar algo mas, para el primer ejemplo decidi poner tal cual una lista de mascotas que le pedi a chatgpt que me hiciera y luego creo una funcion que devuelve un booleano que luego usare con un filter


lista_nueva_prohibidas = list(filter(mascotas_prohibidas, prohibidas))
print("y el resumen de las mascotas prohibidas por aqui abajo... Peligro")
for indice, mascota in enumerate(lista_nueva_prohibidas, start=1):
    print(f'{indice} - {mascota}')
# Y este es el segundo ejemplo que es igual pero diferente jejej, porque aqui reproduzco son las mascotas prohibidas
```

### Tercer ejemplo sin usar filter

```python
mascotas = ["Perro", "Gato", "Mapache", "Loro", "Serpiente Pitón","Hámster", "Cocodrilo", "Conejo", "Oso", "Tortuga","Hurón", "Erizo", "Camaleón", "Tarántula", "Serpiente Coral","Lémur", "Cebra"]
prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]


def mascotas_sin_excluir(lista):
    """Filtra una lista de mascotas, excluyendo las prohibidas.

    Recorre la lista de entrada y conserva solo aquellas mascotas
    que NO están en la lista de prohibidas.

    Args:
        lista (list): Lista de nombres de mascotas a filtrar.

    Returns:
        list: Nueva lista con las mascotas permitidas.
    """
    lista_nueva = []
    for mascota in lista:
        if mascota not in prohibidas:
            lista_nueva.append(mascota)
    return lista_nueva

resultado = mascotas_sin_excluir(mascotas)
for indice, mascota in enumerate(resultado, start=1):
    print(f'{indice} - {mascota}')

# Este es el mismo ejemplo de arriba pero aqui quise mostrarte que se puede hacer el ejercicio sin un filter y bueno tambien queria comerme la cabeza haciendolo
```
