### Enunciado: Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el carácter '#' excepto los últimos cuatro.

### Ejempo con linea de string

```python
datos = "123456789 HolaMundo abc Python secreto abcd"
def convertir_datos_a_caracteres_string(texto):
    """Enmascara las palabras de un string, mostrando solo los últimos 4 caracteres de cada una.

    Separa el texto por espacios, y para cada palabra, si tiene más de 4
    caracteres reemplaza los iniciales con '#', conservando los últimos 4.
    Palabras de 4 o menos caracteres se mantienen intactas.

    Args:
        texto (str): Cadena de texto con palabras separadas por espacios.

    Returns:
        list: Lista de palabras enmascaradas.
    """
    palabras = list(texto.split())
    lista_nueva = []
    for palabra in palabras:
        if len(palabra) <= 4:
            lista_nueva.append(palabra)
        else:
            mascara = "*" * (len(palabra) - 4)
            lista_nueva.append(mascara + palabra[-4:])
    return lista_nueva
resultado = convertir_datos_a_caracteres_string(datos)
print(resultado)
# Creo que esperabas que fuera algo asi el ejercicio porque no supe muy bien cuando decias que tenia que ser una cadena de texto, entonces me imagine en algo asi, pero debia de poder contar el numero de elementos dentro de la cadena de texto entonces por esa razon hice un split y lo converti en una lista para poder trabajar con el mas sencillo
```

### Ejemplo con una lista de texto

```python
datos = ["123456789", "HolaMundo", "abc", "Python", "secreto", "abcd"]
def convertir_datos_a_caracteres(lista):
    """Enmascara todas las palabras de una lista, mostrando solo los últimos 4 caracteres.

    Para cada palabra, si tiene más de 4 caracteres reemplaza los primeros
    caracteres con '#', conservando los últimos 4 visibles. Las palabras
    de 4 o menos caracteres se devuelven sin cambios.

    Args:
        lista (list): Lista de strings a procesar.

    Returns:
        list: Nueva lista con las palabras enmascaradas.
    """
    lista_nueva = []
    for palabra in lista:
        if len(palabra) <= 4:
            lista_nueva.append(palabra)
        else:
            mascara = "*" * (len(palabra) - 4)
            lista_nueva.append(mascara + palabra[-4:])
    return lista_nueva
resultado = convertir_datos_a_caracteres(datos)
print(resultado)
# Para este ejercicio decidí tomar una serie de strings dentro de una lista y poder por un lado guardar las palabras que cumplen la condicion de ser menores a 4 y por el otro he creado una mascara que me permite multiplicar la cantidad de las palabras y formatear esas mismas palabras, menos sus ultimos cuatro caracteres y al final haciendo un append me permite guardar los ultimos cuatro elementos para que se puedan mostrar
```
