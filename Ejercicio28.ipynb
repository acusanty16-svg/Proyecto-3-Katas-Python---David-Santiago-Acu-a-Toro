### Enunciado: Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

### Ejemplo del primer elemento encontrado

```python
numeros = [3, 7, 2, 8, 3, 5, 7, 1, 9, 2]
def contar_elementos(lista):
    """Encuentra el primer número duplicado en una lista.

    Recorre la lista usando un set para recordar los números ya vistos.
    En cuanto encuentra un número que ya está en el set, lo devuelve
    como primer duplicado.

    Args:
        lista (list): Lista de números enteros.

    Returns:
        int or None: El primer número duplicado, o None si no hay duplicados.
    """
    numero_repetido = set()
    for numero in lista:
        if numero in numero_repetido:
            return numero
        numero_repetido.add(numero)
    return None
resultado = contar_elementos(numeros)
print(f'El el pimer numero repetido es: {resultado}')
# Para este ejercicio lo que he hecho es muy parecido al ejercicio 1 con la diferencia de poner un set para simplemente encontrar el primer elemento duplicado y basicamente pregunto dentro un bucle si el elemento ya esta dentro del numero repetido y si es correcto, entonces devulve el primer elemento encontrado
```
