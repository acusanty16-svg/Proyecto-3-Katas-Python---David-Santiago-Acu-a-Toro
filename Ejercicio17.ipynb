### Enunciado: Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2] corresponde al número 572. Usa la función reduce().

### Ejemplo de la funcion reduce

```python
from functools import reduce

lista=[5,7,2]

def funcion_reduce(lista):
    """Convierte una lista de dígitos en el número entero correspondiente.

    Usa reduce() para combinar cada dígito multiplicando el acumulador
    por 10 y sumando el dígito actual. Por ejemplo, [5, 7, 2] → 572.

    Args:
        lista (list): Lista de dígitos (enteros del 0 al 9).

    Returns:
        int: Número entero formado por la concatenación de los dígitos.
    """

    return reduce(lambda acumulador, numero: acumulador * 10 +numero, lista)
resultado = funcion_reduce(lista)
print(f'El valor de la lista {lista} es: {resultado}')
# Este ejercicio intenté hacerlo sin la ayuda de la ia, pero me fue imposible, incluso revisando las notas del master no me quedo del todo claro como usar el reduce de otra forma que esta para realizar el ejercicio, practicaré mas para sacarlo mejor en ejercicios futuros
#Definimos una nueva lista de números
```
