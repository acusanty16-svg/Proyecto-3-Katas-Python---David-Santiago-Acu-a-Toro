# Enunciado: Crea una función que tome un nombre completo y una lista de empleados, busque el nombre en la lista y devuelva el puesto del empleado si se encuentra; de lo contrario, devuelve un mensaje indicando que la persona no trabaja aquí.

### Ejemplo de lista de empleados con un cargo establecido

```python
empleados = [("Michael Scott", "Gerente Regional"),("Dwight Schrute", "Asistente del Gerente Regional"),("Jim Halpert", "Vendedor"),("Pam Beesly", "Recepcionista"),("Stanley Hudson", "Vendedor"),]

def buscar_empleado(lista):
    """Busca un empleado por nombre en una lista de tuplas (nombre, cargo).

    Solicita un nombre al usuario, lo busca en la lista sin distinguir
    mayúsculas/minúsculas. Si lo encuentra, muestra el cargo y termina.
    Si no, pregunta si desea reintentar o salir. Maneja errores de entrada
    con ValueError.

    Args:
        lista (list): Lista de tuplas (nombre_empleado, cargo).

    Returns:
        None: La función imprime resultados y termina con return o break.
    """
    while True:
        try:
            nombre = input("Introduce el nombre del empleado que deseas buscar (o 'salir' para terminar): ")
            if nombre.lower() == "salir":
                print("Saliendo...")
                break
            for empleado,cargo in lista:
                if empleado.lower() == nombre.lower():
                    print(f'{nombre} se encuentra y su cargo es: {cargo}')
                    return
            else:
                    print(f'EL nombre {nombre} no esta en nuestra base de datos')
                    eleccion = int(input("¿Quieres intentarlo nuevamente? introduce '1' para continuar y '2' para salir"))
                    if eleccion == 1:
                        continue
                    #Aqui podria hacer recursividad y llamar a si mismo a la funcion hacer un buscar_empleados(), pero como ya tenia un bucle while no lo vi necesario y simplemente un continue y los otros dos con un return para cortar el juego
                    elif eleccion == 2:
                        print("Saliendo...")
                        return
                    else:
                        print("Introduce un numero para jugar otra vez, Saliendo...")
                        return
        except ValueError:
            print("Introduce un elemento correcto para continuar")

buscar_empleado(empleados)
# Para este ejercicio use algo muy parecido por no decir lo mismo que en el ejercicio30, y basicamente tengo una lista de elementos pares que necesito leer entonces creo un bucle y añado las excepciones para limitar el margen de error y seguidamente busco el nombre del empleado con un input entones el bucle for lee el nombre y si corresponde devuelve el cargo. Pero si no lo encuentra, entonces devuelve un mensaje diciendo que si quiere jugar una vez mas...
```
