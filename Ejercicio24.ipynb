### Enunciado: Calcula la diferencia total en los valores de una lista. Usa la función reduce().

### Ejemplo con lambda

```python
from functools import reduce
numeros = [100, 25, 15, 10, 8, 5]

def resultado_resta(lista):
    """Calcula la resta acumulativa de todos los números en una lista.

    Usa reduce() para restar secuencialmente cada número: acc - numero.

    Args:
        lista (list): Lista de números enteros o flotantes.

    Returns:
        int o float: Resultado de restar todos los números de la lista.
    """

    return reduce(lambda acumulador, numero: acumulador - numero, lista)
resultado = resultado_resta(numeros)
print(f'Dados los numeros: {numeros}\nSu correspondiente resta es: {resultado}')
# Para este ejercicio simplemente he puesto dentro de la funcion una resta de los acumuladores para obtener el resultado de la resta esperada
```

### Ejemplo sin lambda

```python
from functools import reduce
numeros = [100, 25, 15, 10, 8, 5]

def restar (acumulador, numero):
    """Resta dos números.

    Args:
        acumulador (int/float): Valor acumulado inicial.
        numero (int/float): Número a restar.

    Returns:
        int/float: Resultado de la resta.
    """

    return acumulador - numero
def resultado_resta_sin (lista):
    """Calcula la resta acumulativa de todos los números en una lista usando reduce().

    Args:
        lista (list): Lista de números enteros o flotantes.

    Returns:
        int/float: Resultado de restar todos los números secuencialmente.
    """
    return reduce(restar, lista)

resultado = resultado_resta_sin(numeros)
print(f'Dados los numeros: {numeros}\nSu correspondiente resta es: {resultado}')
# Este ejercicio es exactamente igual que el anterior pero sin lambda entonces el codigo se hace un poco mas grande, pero creo que es mas entendible y rapido de hacer con lambda. Basicamente lo que he hecho es crear una funcion que reste el acumulador y el numero, y en la otra una funcion que pueda tomar todos los elementos de la lista, ir acumulandolos y restarlos al final
```
