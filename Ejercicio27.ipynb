### Enunciado: Crea una función que calcule el promedio de una lista de números.

### Ejemplo con tuplas

```python
numeros = [8, 7, 9, 6, 8, 10, 7, 9, 6, 8, 7, 9, 10, 8, 7]
def media(lista):
    """Calcula el promedio de una lista de números usando sum() y len().

    Args:
        lista (list): Lista de números enteros o flotantes.

    Returns:
        float: El promedio de los números en la lista.
    """

    sumatorio = sum(lista)
    cantidad = len(lista)
    resultado = sumatorio / cantidad
    return resultado
resultado = media(numeros)
print(resultado)
# Este es el ejemplo facilito del ejercicio donde solamente con un len y un sum se puede hacer el ejercicio
```

### Ejemplo con reduce

```python
from functools import reduce
def media_reduce(lista):
    """Calcula el promedio de una lista de números usando reduce() para la suma.

    Args:
        lista (list): Lista de números enteros o flotantes.

    Returns:
        float: El promedio de los números en la lista.
    """
    suma_total = reduce(lambda acumulador, numero: acumulador + numero, lista)
    return suma_total / len(lista)
def ejercicio_reduce():
    """Solicita números al usuario, los valida (0-10) y calcula su promedio.

    Pide la cantidad de elementos, luego cada número individualmente.
    Valida que cada número sea <= 10 y que la cantidad sea positiva.
    Usa media_reduce() para calcular el promedio final.

    Returns:
        float: El promedio de los números ingresados.
    """
    while True:
        try: 
            lista_numeros= []
            numeros = int(input("Introduce el numero de elementos que quieres promediar: "))
            if numeros <= 0:
                print("Debe ser un numero positivo")
                continue
            for i in range(numeros):
                while True:
                    numero = int(input(f'Introduce el numero {i+1}:'))
                    if numero <= 10:
                        lista_numeros.append(numero)
                        break
                    print("Tienes que introducir un numero menor o igual a 10 para continuar")
            resultado = media_reduce(lista_numeros)
            print(f'El promedio es: {resultado}')
            return resultado
        except ValueError:
            print("Error al compilar la operacion porque has hecho al mal")
ejercicio_reduce()
# En cambio este es el ejercicio tocho donde tuve que dividir el ejercicio en dos partes porque creo que de otra forma hubiera sido imposible hacerlo, y en una funcion hago la operacion del enunciado con un reduce que es menos codigo y en el segundo tenia que manejar tanto excepciones como inputs para que el que quiera entrar a hacer algo no haga lo que quiera y rompa el programa. Es muy parecido al ejercicio 10 donde tambien tenia que manejar excepciones para optener un promedio.
```
