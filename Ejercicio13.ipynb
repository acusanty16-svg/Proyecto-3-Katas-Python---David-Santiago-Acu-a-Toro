# Decimotercera Kata

### Enunciado: Genera una función que, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas. Usa la función map().

### Ejemplo de texto y mapeo de dicho texto

```python
texto = "aBbcC DeFfG Gh" 
nuevo_texto = set(texto.lower().replace(" ","" ))
def formatear_caracter(caracter):
    """Convierte un carácter en una tupla con su versión mayúscula y minúscula.

    Args:
        caracter (str): Un único carácter.

    Returns:
        tuple: Tupla (mayúscula, minúscula) del carácter.
    """
    return (caracter.upper(), caracter.lower())
resultado = list(map(formatear_caracter,nuevo_texto))
print(resultado)
# Para este ejercicio lo que he usado es un seteador del texto original y seguidamente he creado un funcion que devuelve una tupla del caracter tanto en mayus como en minus y al final he mapeado una lista nueva que contiene cada una de los elementos sin repetir, y por esa razon he decidido setearlo para que no se repita ninguna palabra dentro de las tuplas incluso si existiese algun espacio entre ellas, porque con el replace " " por "" no va a contar los espacios
```
