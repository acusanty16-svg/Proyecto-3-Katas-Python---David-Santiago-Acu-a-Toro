# Decimonovena Kata

### Enunciado: Crea una función lambda que filtre los números impares de una lista dada.

### Ejermplo con dos funciones que devulve tanto numeros pares como impares

```python
numeros = [12, 7, 25, 8, 31, 14, 19, 10, 5, 22, 33, 16, 41, 6, 27, 18, 9, 30, 45, 2]

def devolver_impares(lista):
    """Filtra los números impares de una lista usando filter().

    Args:
        lista (list): Lista de números enteros.

    Returns:
        list: Nueva lista con solo los números impares.
    """

    return list(filter(lambda numero: numero % 2 != 0, lista))

def devolver_pares(lista):
    """Filtra los números pares de una lista usando filter().

    Args:
        lista (list): Lista de números enteros.

    Returns:
        list: Nueva lista con solo los números pares.
    """

    return list(filter(lambda numero: numero % 2 == 0, lista))

resultado_impar = devolver_impares(numeros)
resultado_par = devolver_pares(numeros)

print(f'tenemos la siguiente lista: {numeros}\nSus correspondientes numeros impares son: {resultado_impar} \nSus pares son: {resultado_par}')
# Para este ejercicio he decidido tanto mostrar los numeros pares como los impares dentro de dos funciones porque creo que se veria mas interesante y podemos tener un panorama mas completo
```
