### Enunciado: Crea una función llamada procesar_texto
Procesa un texto según la opción especificada: contar_palabras, reemplazar_palabras o eliminar_palabra.
Código a seguir:
Crear una función contar_palabras que cuente el número de veces que aparece cada palabra en el texto y devuelva un diccionario.
Crear una función reemplazar_palabras para sustituir una palabra_original por una palabra_nueva en el texto y devolver el texto modificado.
Crear una función eliminar_palabra que elimine una palabra del texto y devuelva el texto sin ella.
Crear la función procesar_texto que reciba un texto, una opción ("contar", "reemplazar", "eliminar") y un número variable de argumentos según la opción elegida.
Caso de uso:
Verificar el funcionamiento completo de procesar_texto.

### Ejemplo con el texto propuesto

```python
texto = "Michael Scott es el mejor gerente que ha tenido Dunder Mifflin y tambien es el peor gerente que ha tenido Dunder Mifflin pero todos lo queremos porque Michael es Michael"

def contar_palabras(texto):
    print("Bienvenido a contar palabras!!")
    nuevo_dict={}
    for palabra in texto.split():
        palabra = palabra.lower()
        if palabra in nuevo_dict:
            nuevo_dict[palabra] += 1
        else:
            nuevo_dict[palabra] = 1
    for palabra, cantidad in nuevo_dict.items():
        print(f'la palabra: "{palabra}" aparece {cantidad} veces')
    print()
    return nuevo_dict
    #Este primer metodo me devuelve el numero de palabras que hay dentro de un texto introducido

def reemplazar_palabras(texto):
    print("Bienvenido a reemplazar palabras")
    original = input("Introduce la palabra original: ")
    nueva = input("Introuce la palabra a reemplazar: ")
    palabras = texto.split()
    for i, palabra in enumerate(palabras):
        if palabra.lower() == original.lower():
            palabras[i] = nueva
    palabras = " ".join(palabras)
    print(f'La palabra original es: "{original}" y la palabra a reemplazar es: "{nueva}"\nSu nuevo string es: {palabras}')
    print()
    return palabras
    #Para este metodo necesite ayuda al final con el join porque al final no me salia un string sino como una lista y pues por ahi no es el trabajo

def eliminar_palabra(texto):
    print("Bienvenido a eliminar palabra")
    eliminar= input("Introduce la palabra que quieres eliminar: ")
    palabras = texto.split()
    palabras_filtradas = []
    for palabra in palabras:
        if palabra.lower() != eliminar.lower():
            palabras_filtradas.append(palabra)
    palabras = " ".join(palabras_filtradas)
    print(f'La palabra que desea eliminar es: "{eliminar}"\nSu nuevo string es: {palabras}')
    print()
    return " ".join(palabras_filtradas)
    #Este metodo introduce una palabra con un input y seguidamente los filtro mediante un lower para que lee la palabra en minuscula y no haga excepcion de ninguna, despues le digo que si cualquier palabra es diferente de la palabra eliminar entonces que la guarde dentro de una lista a la cual debo formatear para volverla un string y devolverla al final del metodo

def procesar_texto(texto, opcion):
    try:
        if opcion == "contar":
            return contar_palabras(texto)
        elif opcion == "reemplazar":
            return reemplazar_palabras(texto)
        elif opcion == "eliminar":
            return eliminar_palabra(texto)
        else:
            raise ValueError("Opcion no valida")
    except (ValueError, TypeError):
        print("Introduce la informacion valida correspondiente para poder continuar")
    #Y este es el metodo mas facil jejej, porque aqui solamente llamo a los otros metodos y me deja trabajar con ellos tambien, entonces es el mas accesible con menos fallo. de igual manera quise poner un try para asegurarnos de cualquier posible error


procesar_texto(texto, "contar")
procesar_texto(texto, "reemplazar")
procesar_texto(texto, "eliminar")
procesar_texto(texto,"Hola_Amigos")

#Vale fue necesario actualizar todo el codigo para este punto, porque reproducia todos los metodos pero no tenia mucho sentido lo que hacia porque solamente botada resultado sin sentido, entonces he tenido que agregar dentro de cada metodo su correspondiente ejecucion dentro de varios prints para que tuviera un poco mas de logica
```
