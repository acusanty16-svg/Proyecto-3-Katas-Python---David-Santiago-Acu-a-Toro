# Trigésimo séptima Kata

### Enunciado: Genera un programa que nos indique si es de noche, de día o de tarde según la hora proporcionada por el usuario.

### Ejemplo de horario habitual

```python
def momento_del_dia():
    """Determina el momento del día según la hora ingresada por el usuario.

    Solicita una hora (0-23) en un bucle continuo hasta que el usuario
    ingrese un valor válido. Según el rango horario, retorna un mensaje
    indicando si es mañana, tarde, noche o madrugada.

    Returns:
        str: Mensaje con el momento del día correspondiente a la hora ingresada.
    """
    while True:
        try:
            hora = int(input("Introduce la hora (0-23): "))
            if hora < 0 or hora > 23:
                print("Hora invalida. Debe ser entre 0 y 23")
                continue
            if 6 <= hora <= 11:
                return "Buenos dias, al que madruga Dios lo ayuda"
            elif 12 <= hora <= 18:
                return "Buenas tardes, ya has almorzao?"
            elif 19 <= hora <= 23:
                return "Buenas noches, a mimir que mañana hay que levantar España"
            else:
                return "Buenas noches, es de madrugada"
        except ValueError:
            print("Introduce un numero valido para continuar")
    #Para este ejercicio lo que he decidido hacer es un if con diferentes opciones dependiendo del numero que introduzca la persona, y tambien he manejado los try para que podamos controlar mas los errores

print(momento_del_dia())
```
