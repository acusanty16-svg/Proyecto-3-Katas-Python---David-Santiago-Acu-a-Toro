### Enunciado: Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función map().

### Ejemplo de tupla con nombre y edad

```python
datos = [("Santiago", 25), ("Andrea", 30), ("Carlos", 22)]

def tupla_string(tupla):
    """Formatear el nombre y edad de una tupla

    Args:
        tupla (tuple): Recibes una tupla

    Returns:
        str: Devuelve el nombre y la edad formateados de dicha tupla
    """
    nombre = tupla[0]
    edad = tupla[1]
    return f'{nombre} tiene {edad} años'

resultado = list(map(tupla_string,datos))
print (resultado)

## para este ejercicio he decidido hacerlo en referencia al nombre y edad de individuos, pero si se quisiese hacer para algo mas, solamente tendria que poner la posicion dentro de la tupla y de esa misma forma formatearla para que el argumento sea un string
```
