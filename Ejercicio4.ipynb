### Único ejemplo

```python
lista_a = [10, 25, 7, 3, 6, 9, 22, 5, 13, 12]
lista_b = [3, 8, 2, 11, 5, 7, 15, 4, 13, 9]

def diferencia (listaa,listab):
    """La diferencia entre dos listas

    Args:
        listaa (list): elementos de la primera lista
        listab (list): elementos de la segunda lista

    Returns:
        list: Devuelve la diferencia entre los listas
    """
    print(f'{listaa} - {listab} = {listaa - listab}')
    return listaa-listab
resultado = list(map(diferencia,lista_a,lista_b))

# Para este ejercicio solamente pude pensar en una forma de realizarlo, que era la mas facil ejejj, solamente restando las dos listas, porque intente hacerlo de diferentes formas pero el resultado no era el esperado 
```
