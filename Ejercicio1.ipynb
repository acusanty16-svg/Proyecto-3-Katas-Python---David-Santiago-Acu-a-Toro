Enunciado: Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def diccionario_con_frecuencia(texto):
    """Diccionario con frecuencia

    Args:
        texto (dict): Introduces las palabras que quieras revisar

    Returns:
        dict: Devuelve la frecuencia de cada caracter y hace un continue cuando hay un espacio, una coma o un punto
    """
    diccionario_nuevo = {}
    for letra in texto:
        if letra in " ,.":
            continue
        letra = letra.lower()
        if letra in diccionario_nuevo:
            diccionario_nuevo[letra] += 1
        else:
            diccionario_nuevo[letra] =1
    return diccionario_nuevo
        
diccionario_con_frecuencia("Mi mama me mima, como me mima, mi mama")

# Aqui opte por hacer una funcion que al introducir una serie de parametros strings, los lea y si encuentra tanto una coma, como un punto o como un espacio tendra que hacer un continue, luego contará y sumará cada letra dentro del nuevo diccionario y así saber cuantas letras tiene cada uno... Tuve que usar un lower() porque sino aparecian todas las caracteristicas de upperCase
    
            
```

```python
def diccionario_con_frecuencia2(*args):
    """diccionario con frecuencias pero recibiendo un args indefinido

    Args:
        texto (str): Introduces las palabras que quieras revisar
    Returns:
        dict: Devuelve la frecuencia de cada caracter y hace un continue cuando hay un espacio, una coma o un punto
    """
    diccionario_nuevo = {}
    for texto in args:
        for letra in texto:
            if letra in " ,.":
                continue
            letra = letra.lower()
            if letra in diccionario_nuevo:
             diccionario_nuevo[letra] +=1
            else:
                diccionario_nuevo[letra] = 1
    return diccionario_nuevo

diccionario_con_frecuencia2("Mi mama me mima, como me mima, mi mama")

# la idea es la misma que en el anterior ejemplo pero la diferencia radica en el tipo de parametro que lee python, en este caso va a leer una serie de caracteres, pero es necesario anidar esos caracteres dentro de otro bucle para poder leer letra a letra, sino va a salir un solo resultado
```
