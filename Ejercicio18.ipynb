### Enunciado: Escribe un programa en Python que cree una lista de diccionarios con información de estudiantes (nombre, edad, calificación) y use filter para extraer a los estudiantes con una calificación mayor o igual a 90.

### Ejemplo con filter como lo pide el enunciado

```python
estudiantes = [
    {"nombre": "Hermione", "edad": 17, "calificacion": 98},
    {"nombre": "Ron", "edad": 17, "calificacion": 72},
    {"nombre": "Harry", "edad": 17, "calificacion": 85},
    {"nombre": "Luna", "edad": 16, "calificacion": 95},
    {"nombre": "Draco", "edad": 17, "calificacion": 88},
    {"nombre": "Neville", "edad": 16, "calificacion": 91},
    {"nombre": "Ginny", "edad": 16, "calificacion": 94},
    {"nombre": "Cedric", "edad": 17, "calificacion": 90},
]
def tener_estudiantes(lista):
    """Filtra los estudiantes con calificación mayor o igual a 90.

    Utiliza filter() con una lambda para conservar solo aquellos estudiantes
    cuya calificación sea >= 90 de la lista de diccionarios.

    Args:
        lista (list): Lista de diccionarios de estudiantes, cada uno con
                      las claves "nombre", "edad" y "calificacion".

    Returns:
        list: Nueva lista con los estudiantes que cumplen la condición.
    """

    return list(filter(lambda estudiante: estudiante["calificacion"]>=90, lista))
tener_estudiantes(estudiantes)
# Para este ejercicio he usado el filter tal y como indicaba el enunciado, entonces lo unico que hice fue buscar el apartado especifico de cada estudiante donde estuviera registrada la calificacion e indique con un condicional si era mayor o igual a 90
```

### Ejemplo extra sin el filter

```python
estudiantes = [
    {"nombre": "Hermione", "edad": 17, "calificacion": 98},
    {"nombre": "Ron", "edad": 17, "calificacion": 72},
    {"nombre": "Harry", "edad": 17, "calificacion": 85},
    {"nombre": "Luna", "edad": 16, "calificacion": 95},
    {"nombre": "Draco", "edad": 17, "calificacion": 88},
    {"nombre": "Neville", "edad": 16, "calificacion": 91},
    {"nombre": "Ginny", "edad": 16, "calificacion": 94},
    {"nombre": "Cedric", "edad": 17, "calificacion": 90},
]
def tener_estudiantes(lista):
    """Filtra los estudiantes con calificación mayor o igual a 90.

    Itera sobre la lista de diccionarios de estudiantes y conserva
    aquellos cuya calificación sea >= 90.

    Args:
        lista (list): Lista de diccionarios de estudiantes, cada uno con
                      las claves "nombre", "edad" y "calificacion".

    Returns:
        list: Nueva lista con los estudiantes que cumplen la condición.
    """

    lista_nueva = []
    for estudiante in lista:
        if estudiante["calificacion"] >=90:
            lista_nueva.append(estudiante)
    return lista_nueva

tener_estudiantes(estudiantes)
# Y aqui pense en que se podria plantear el mismo ejercicio pero de otra manera para ver que sin el lambda o sin el filter tambien se puede realizar el mismo ejercicio
```
