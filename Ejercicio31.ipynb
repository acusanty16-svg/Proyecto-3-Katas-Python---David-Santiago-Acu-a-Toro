### Enunciado: Crea una función que solicite al usuario ingresar una lista de nombres y luego un nombre para buscar en esa lista. Si el nombre está en la lista, imprime un mensaje indicando que fue encontrado; de lo contrario, lanza una excepción.

### Ejemplo de busqueda de empleado

```python
# Para hacer una excepcion personalizada me ayudo mi fiel amigo opencode pero si es verdad que el resto del codigo intente hacerlo yo solo sin ayuda y tratando de error en error hasta encontrar la manera mas facil de arreglarlo todo
class NombreNoencontradoError(Exception):
    pass

def introducir_empleado():
    """Solicita al usuario una lista de nombres de empleados.

    Pide la cantidad de empleados y luego cada nombre individualmente.
    Valida que la cantidad sea un número positivo y maneja errores de
    entrada con ValueError.

    Returns:
        list: Lista con los nombres de los empleados ingresados.
    """
    while True:
        try:
            lista_empleados = []
            numero_empleados= int(input("Introduce el numero de empleados: "))
            if numero_empleados <= 0:
                print("Introduce un numero positivo de empleados para continuar")
                continue
            else:
                for i in range (numero_empleados):
                    nombre_empleado= input(f'Introduce el nombre del empleado {i+1}: ')
                    lista_empleados.append(nombre_empleado)
            return lista_empleados
        except ValueError:
            print("El numero de empleados introducidos es incorrecto")

def buscar_empleado(lista):
        """Busca un nombre en la lista de empleados y lanza excepción si no existe.

    Solicita al usuario el nombre a buscar, recorre la lista y si lo
    encuentra imprime un mensaje de éxito. Si no está en la lista,
    lanza una excepción NombreNoencontradoError.

    Args:
        lista (list): Lista de nombres de empleados.

    Returns:
        bool or None: True si encuentra el nombre, o lanza excepción si no.

    Raises:
        NombreNoencontradoError: Si el nombre no está en la lista.
    """
        nombre_empleado = input("Introduce el nombre del empleado a buscar: ")
        for empleado in lista:
            if empleado == nombre_empleado:
                print(f'Empleado {nombre_empleado} encontrado dentro de la base de datos')
                return True
        raise NombreNoencontradoError(f'El nombre {nombre_empleado} no esta en la lista')

try:
    empleados = introducir_empleado()
    buscar_empleado(empleados)
except NombreNoencontradoError as mensaje:
    print(mensaje)
# Pues sentimientos encontrados en este ejercicio, uno es que quise hacerlo un poco mas complicado y enrevesado y quizas hubiera podido haberlo hecho más facil, pero quise intentar cosas nuevas como la excepcion nueva pero el resto del codigo si que ha sido labor mia, hubo un pequeño dilema que no me permitia jugar bien con el codigo y no iteraba muy bien y fue en el bucle del for porque si no pongo el range no itera y aparecia error
```
