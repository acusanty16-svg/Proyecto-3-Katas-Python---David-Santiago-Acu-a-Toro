Enunciado: Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def buscar_objetivo(lista, objetivo):
    """Encontrar el string de la lista

    Args:
        objetivo (str): Introduces tanto la lista como el string que deseas buscar

    Returns:
        list: Devuelve el indice donde se encuentra el elemento, empezando en 1 y añade los elementos a la nueva lista.

    """
    nueva_lista = []
    for indice, letra in enumerate(lista):
        if objetivo in letra:
            nueva_lista.append(letra)
            print(f'Se ha encontrado la palabra {objetivo} dentro de la lista con el indice: {indice}')
    return nueva_lista
buscar_objetivo(["varita", "escoba", "horrocrux", "sombrero", "cicatriz","hogwarts", "horrocrux", "mapa", "capas", "pocima","biblioteca", "dragón", "basiliscos", "horrocrux", "espejo","diario", "horrocrux", "murciélago", "caldero", "lechuza"], "horrocrux")
```
